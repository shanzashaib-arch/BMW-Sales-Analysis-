```python
# Step 1: Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Step 2: Upload Excel File
uploaded = files.upload()  # File chooser open hoga, apni Excel file select karo

# Step 3: Load Excel Dataset
# Uploaded file ka exact name copy karo yahan
df = pd.read_excel("BMW sales data (2010-2024) (1).csv.xlsx")

# Step 4: Quick Look at Data
print("Dataset Shape:", df.shape)
print(df.head())

# Step 5: Check for Missing Values
print("\nMissing Values:\n", df.isnull().sum())

# Step 6: Basic Statistics
print("\nBasic Stats:\n", df.describe())

# Step 7: Data Cleaning
# Agar Date column hai to usko datetime me convert karo
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])

# Step 8: Example Analysis - Total Sales per Year
if 'Year' in df.columns and 'Sales' in df.columns:
    sales_per_year = df.groupby('Year')['Sales'].sum()
    print("\nTotal Sales Per Year:\n", sales_per_year)

    # Visualization
    plt.figure(figsize=(12,6))
    sales_per_year.plot(kind='bar', color='skyblue')
    plt.title("BMW Total Sales Per Year (2010-2024)")
    plt.xlabel("Year")
    plt.ylabel("Total Sales")
    plt.grid(axis='y')
    plt.show()
else:
    print("Dataset me 'Year' ya 'Sales' column missing hai. Column names check karo:", df.columns)

# Step 9: Optional - Save Cleaned Data
df.to_excel("cleaned_BMW_sales.xlsx", index=False)
print("Cleaned dataset saved as 'cleaned_BMW_sales.xlsx'")
```
